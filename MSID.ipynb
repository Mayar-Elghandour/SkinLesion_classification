```python
import pandas as pd
import cv2
from os import listdir
import matplotlib.pyplot as plt
```

```python
chickenpox_path="/kaggle/input/monkeypoxskinimagedataset/Monkeypox Skin Image Dataset/Chickenpox"
measles_path="/kaggle/input/monkeypoxskinimagedataset/Monkeypox Skin Image Dataset/Measles"
monkeypox_path="/kaggle/input/monkeypoxskinimagedataset/Monkeypox Skin Image Dataset/Monkeypox"
normal_path="/kaggle/input/monkeypoxskinimagedataset/Monkeypox Skin Image Dataset/Normal"
```

```python
MSID_1=pd.DataFrame()
MSID_1["filename"]=listdir(monkeypox_path)
MSID_1["paths"]=monkeypox_path+"/"+MSID_1["filename"]
MSID_1["labels"]="mpox"
MSID_1
```

```python
MSID_2=pd.DataFrame()
MSID_2["filename"]=listdir(chickenpox_path)
MSID_2["paths"]=chickenpox_path+"/"+MSID_2["filename"] 
MSID_2["labels"]="chpox"
MSID_2
```

```python
MSID_3=pd.DataFrame()
MSID_3["filename"]=listdir(measles_path)
MSID_3["paths"]=measles_path+"/"+MSID_3["filename"]
MSID_3["labels"]="measles"
MSID_3
```

```python
MSID_4=pd.DataFrame()
MSID_4["filename"]=listdir(normal_path)
MSID_4["paths"]=normal_path+"/"+MSID_4["filename"]
MSID_4["labels"]="normal"
MSID_4
```

```python
MSID=pd.DataFrame()
MSID=pd.concat([MSID_1,MSID_2, MSID_3,MSID_4])
MSID
```

```python
MSID.to_csv('MSID.csv', index=False) 
```

visualisation and data analysis

```python
MSID["labels"].value_counts()
```

```python
MSID["labels"].value_counts().plot.pie(figsize=(6, 6),autopct='%1.1f%%')
```

```python
example=MSID.groupby("labels").apply(lambda x: x.sample(1),include_groups=False)
example
```

```python
plt.figure(figsize=(6,6))
for i,paths in enumerate(example["paths"]):
    image=cv2.imread(paths)
    plt.subplot(2,2,i+1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(example.index[i][0])
```
