```python
import os
import pandas as pd 
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
data = list()
for root, dirs, files in os.walk('/kaggle/input/mpox-skin-lesion-dataset-version-20-msld-v20/Original Images/Original Images/FOLDS/fold1'):
    for filename in files:
        if "CHP" in filename:
            fullpath = os.path.join(os.path.abspath(root),filename)
            data.append((filename, fullpath, 'chpox'))
        elif "CWP" in filename:
            fullpath = os.path.join(os.path.abspath(root),filename)
            data.append((filename,fullpath,'CWP'))
        elif "HFMD" in filename:
            fullpath = os.path.join(os.path.abspath(root),filename)
            data.append((filename,fullpath,"HFMD"))
        elif "HEALTHY" in filename:
            fullpath = os.path.join(os.path.abspath(root),filename)
            data.append((filename,fullpath,"normal"))
        elif "MSL" in filename:
            fullpath = os.path.join(os.path.abspath(root),filename)
            data.append((filename,fullpath,"measles"))
        else:
            fullpath = os.path.join(os.path.abspath(root),filename)
            data.append((filename,fullpath,"mpox"))
df = pd.DataFrame(data, columns=['filename', 'paths', 'labels'])
print(df)
```

```python
csvFile = df.to_csv('MSLD.csv', index = False) 
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import PIL

# Initialize variables
a, b, c, d, s, k, count = 0, 0, 0, 0, 0, 0, 0

plt.figure(figsize=(8, 8))
for name, paths, col in zip(df['filename'],df['paths'], df['labels']):
    if col == 'chpox' and a == 0:
        print(name,paths,'1\n')
        plt.subplot(3, 2, 1)
        plt.imshow(plt.imread(paths))
        plt.title(col)
        a = 1
        count += 1
    elif col == 'mpox' and b == 0:
        print(name,paths,'2\n')
        plt.subplot(3, 2, 2)
        plt.imshow(plt.imread(paths))
        plt.title(col)
        b = 1
        count += 1
    elif col == 'measles' and c == 0:
        print(name,paths,'3\n')
        plt.subplot(3, 2, 3)
        plt.imshow(plt.imread(paths))
        plt.title(col)
        c = 1
        count += 1
    elif col == 'CWP' and d == 0:
        print(name,paths,'4\n')
        plt.subplot(3, 2, 4)
        plt.imshow(plt.imread(paths))
        plt.title(col)
        d = 1
        count += 1
    elif col == 'HFMD' and s == 0:
        print(name,paths,'5\n')
        plt.subplot(3, 2, 5)
        plt.imshow(plt.imread(paths))
        plt.title(col)
        s = 1
        count += 1
    elif col == 'normal' and k == 0:
        print(name,paths,'6\n')
        plt.subplot(3, 2, 6)
        plt.imshow(plt.imread(paths))
        plt.title(col)
        k = 1
        count +=1
    if count == 6:
        break

plt.tight_layout()
plt.show()
```
